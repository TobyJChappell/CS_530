#### Part a

a. Implement a function that takes in the R<sup>2</sup> of a regression and return the VIF value. 

```python
def calculate_vif(r_squared):
    vif = 1/(1-r_squared)
    return vif
```

#### Part b

```python
np.random.seed(104)
# Dataset 1: (X1, y1)
X1 = np.random.normal(size=(100,10))
y1 = 1*X1[:,0] + 2*X1[:,1] + 3*X1[:,2] + 4*X1[:,3] + 5*X1[:,4] + np.random.normal(0, 1, 100)

# Dataset 2: (X2, y2)
X2 = np.random.normal(size=(100,10))
y2 = 1*X2[:,0] + 2*X2[:,1] + 3*X2[:,2] + 4*X2[:,3] + 5*X2[:,4] + np.random.normal(0, 1, 100)

# Dataset 3: (X3, y3)
X3_ = np.random.normal(size=(100,10))
X_extra = X3_[:,0] + np.random.normal(0,0.2,100)
X_extra = X_extra.reshape((100,1))
X3 = np.hstack([X3_, X_extra])
y3 = 1*X3[:,0] + 2*X3[:,1] + 3*X3[:,2] + 4*X3[:,3] + 5*X3[:,4] + np.random.normal(0, 1, 100)

# Dataset 4: (X4, y4)
X4_ = np.random.normal(size=(100,10))
X_extra = X4_[:,0] + np.random.normal(0,0.2,100)
X_extra = X_extra.reshape((100,1))
X4 = np.hstack([X4_, X_extra])
y4 = 1*X4[:,0] + 2*X4[:,1] + 3*X4[:,2] + 4*X4[:,3] + 5*X4[:,4] + np.random.normal(0, 1, 100)
```

```python
linear4 = sm.OLS(y4,sm.add_constant(X4)).fit()
linear4.summary()
```
