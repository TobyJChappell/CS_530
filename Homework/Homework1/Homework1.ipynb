During class, when we discussed the OLS method for linear regression, we derived $\hat\beta_0$ that minimizes the RSS by taking the partial derivative of the residual sum of squares (RSS) with respective to $\hat\beta_0$. 

$$
\begin{align}
RSS &= \sum^n_{i=1}(y_i-\hat{y_i})^2 \\
&= \sum^n_{i=1}(y_i-\hat\beta_0-\hat\beta_1x_i)^2
\end{align}
$$

Go ahead and similarly derive $\hat\beta_1$ that minimizes the RSS. Show all your derivation steps; only showing the final answer will not be accepted as the correct answer. However, for now, you can also derive it on a piece of paper and upload a photo copy of it to Canvas while you are submitting your homework.

$$
\begin{align}
0 &= \frac{\partial (RSS)}{\partial \hat\beta_1} \\
&= \frac{\partial(\sum^n_{i=1}(y_i-\hat\beta_0-\hat\beta_1x_i)^2)}{\partial \hat\beta_1} \\
&= -2\sum^n_{i=1}x_i(y_i-\hat\beta_0-\hat\beta_1x_i) \\
&= \sum^n_{i=1}x_i(y_i-\hat\beta_0-\hat\beta_1x_i) \\
&= \sum^n_{i=1}x_i(y_i-\bar{y}+\hat\beta_1\bar{x}-\hat\beta_1x_i) \\
&= \sum^n_{i=1}(x_iy_i-x_i\bar{y}+\hat\beta_1x_i\bar{x}-\hat\beta_1x_i^2) \\
&= \sum^n_{i=1}(x_iy_i-x_i\bar{y})-\hat\beta_1\sum^n_{i=1}(x_i^2-x_i\bar{x}) \\
=> \hat\beta_1 &= \frac{\sum^n_{i=1}(x_iy_i-x_i\bar{y})}{\sum^n_{i=1}(x_i^2-x_i\bar{x})} \\
Since \sum^n_{i=1}(\bar{x}\bar{y}-y_i\bar{x}) &= \sum^n_{i=1}(\bar{x}^2-x_i\bar{x}) = 0,\\
=> \hat\beta_1 &= \frac{\sum^n_{i=1}(x_iy_i-x_i\bar{y})+\sum^n_{i=1}(\bar{x}\bar{y}-y_i\bar{x})}{\sum^n_{i=1}(x_i^2-x_i\bar{x})+\sum^n_{i=1}(\bar{x}^2-x_i\bar{x})} \\
&= \frac{\sum^n_{i=1}(x_i-\bar{x})(y_i-\bar{y})}{\sum^n_{i=1}(x_i-\bar{x})^2} \\ \\
\end{align}
$$

* Question2:

Based on the derivation you have from the above for $\hat\beta_0$ and $\hat\beta_1$, design and code functions that calcuate $\hat\beta_0$ and $\hat\beta_1$ from scratch in Python. (Do not just call a Python function that claculates this for you.)

```
import numpy as np
def coeff_linear(X,y):
  temp1=0
  temp2=0
  avgX = sum(X)/len(X)
  avgy = sum(y)/len(y)
  for i in range(len(X)): 
    temp1 = temp1 + (X[i]-avgX)*(y[i]-avgy)
  for i in range(len(X)): 
    temp2 = temp2 + (X[i]-avgX)**2
  beta_1 = temp1/temp2
  beta_0 = avgy - beta_1*avgX
  return beta_0, beta_1
```
